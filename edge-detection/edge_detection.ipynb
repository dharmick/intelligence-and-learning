```python
from PIL import Image
import matplotlib.pyplot as plt
import math
```

```python
img = Image.open("Bugs.jpg")
width, height = img.size
plt.imshow(img)
```

```python
newimg = Image.new("RGB", (width, height), "white")
plt.imshow(newimg)
```

```python
filter1 = [
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
]

filter2 = [
    [-1, -2, -1],
    [0, 0, 0],
    [1, 2, 1]
]
```

```python
for x in range(1, width-1):
    for y in range(1, height-1):
        
        dx = 0
        dy = 0
        
        for i in range(-1,2):
            for j in range(-1,2):
                p = img.getpixel((x-i, y-j))
                r = p[0]
                g = p[1]
                b = p[2]
                dx += filter1[j+1][i+1] * (r + g + b)
                dy += filter2[j+1][i+1] * (r + g + b)

        len = math.sqrt(dx * dx + dy * dy)
        len = len / 4328 * 255
        len = int(len)
        
        newimg.putpixel((x,y),(len,len,len))
```

```python
plt.imshow(newimg)
```
